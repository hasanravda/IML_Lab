# Logistic Regression from Scratch

```python
import numpy as np
```

```python
class Logistic_Regression():
    
    def __init__(self,learning_rate=0.01,no_of_iterations=100):
        self.learning_rate=learning_rate
        self.no_of_iterations=no_of_iterations
        
    def fit(self,X,Y):
        
        # number of data points (no. of rows) --> m
        # number of input fetures (no. of columns) --> n
        self.m,self.n=X.shape
        
        self.w=np.zeros(self.n)
        self.b=0
        
        self.X=X
        self.Y=Y
        
        for i in range(self.no_of_iterations):
            self.update_weights()
        
    
    def update_weights(self):
        
        # y_hat or y_cap value is the probality that value of y will be 1
        # Y_hat formula (Sigmoid Function)  
        
        z=self.X.dot(self.w) + self.b
        Y_hat= 1/( 1 + np.exp (- (self.X.dot(self.w) + self.b) ) )
        
        # derivatives
        
        dw= (1/self.m) * np.dot(self.X.T,(Y_hat - self.Y))  # We are taking transpose of X , so that we can do multiply it with Y
        
        db= (1/self.m) * np.sum(Y_hat - self.Y)
        
        
        # Updating weights and bias using gradient descent algorithm 
        
        self.w = self.w - self.learning_rate * dw 
        self.b =self.b - self.learning_rate * db
        
        
    # Sigmoid Equation & Decision Boundary
    def predict(self,X):
        
        Y_pred = 1/( 1 + np.exp (-(X.dot(self.w) + self.b) ) )
        Y_pred = np.where( Y_pred > 0.5 ,1,0)
        return Y_pred
```

```python
X_train,X_test,Y_train,Y_test = train_test_split (X,y,test_size=0.2,random_state=2)

classifier=Logistic_Regression(learning_rate=0.01,no_of_iterations=100)
classifier.fit(X_train,Y_train)
Y_pred=classifier.predict(X_test)
accuracy_score(Y_test, Y_pred)
```
