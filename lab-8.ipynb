## Write a Pandas program to create a line plot of the opening, closing stock prices of given company between two specific dates.

```python
# Hasan Ravda        216270307028

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('/kaggle/input/stock-market-data-nifty-50-2000-2022/Nifty_50_with_indicators_.csv')
df = df.drop(df.columns[6:],axis = 1)
```

```python
df.head()
```

```python
df_10=df.iloc[0:10]
df_10
```

```python
df=df.iloc[:10]
plt.figure(figsize=(13,5))
plt.plot(df['Date'],df['open'],label='Opening Price', color='blue')
plt.plot(df['Date'],df['close'],label='Closing Price', color='red')
plt.title('Opening and Closing Stock prices')
plt.legend()
```

```python
print("Nifty 50 \n10 days")
df_10.plot(x='Date',y=['open','close'])
# df_100.plot()
```

```python
print('Moving Average for 5 close price ')
df_10.plot(x='Date',y='sma5',ylabel='Simple Moving Average (5days)')
```
